```python
from ktrain import text
from sklearn.datasets import fetch_20newsgroups
```

```python
remove = ('headers', 'footers', 'quotes')

newsgroups_train = fetch_20newsgroups(subset='train', remove=remove)
newsgroups_test = fetch_20newsgroups(subset='test', remove=remove)


docs = newsgroups_train.data + newsgroups_test.data
docs[:100]
```

```python
INDEX_DIR = "tmp/myindex"
```

```python
text.SimpleQA.initialize_index(INDEX_DIR)
```

```python
text.SimpleQA.index_from_list(docs, INDEX_DIR, commit_every=len(docs),
                              multisegment=True, procs=4, 
                              breakup_docs=True)
```

```python
qa = text.SimpleQA(INDEX_DIR)
```

```python
answers = qa.ask("What is fossil fuel plants?")
```

```python
answers
```

```python
qa.display_answers(answers)
```

```python
def ask_ques(query):
  answers = qa.ask(query)
  return qa.display_answers(answers)
```

```python
ask_ques(query="Who was Muhammad?")
```
